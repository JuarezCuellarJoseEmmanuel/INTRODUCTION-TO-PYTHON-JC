## If-else statement
A Python if else statement helps programmers control the flow of their programs by evaluating whether an expression is true or false. If the condition is true, the if statement is executed. Otherwise, the else statement is executed [(CAREER KARMA)](https://careerkarma.com/blog/python-if-else/).

* ### Example 1
the variables x and s are declared and a value is assigned, the if statement is used and an operator symbol is assigned to the greater of which the print() function is used to print if the value is greater and an else is placed with its respective print() to be able to print when it is not greater.

```python
x = 50
s = 10

if x > s:
 print("is greater")

else:
     print("is less")
```

* ### Example 2
The variable R is declared, a value is assigned to it, the if statement is used and a greater than operator symbol is assigned, the print() function is used to print if there is an obstacle nearby and an else with its respective print() is set to know if it is clear.

```python
R = 6
if R >= 6:
  print("There is an obstacle nearby") 
else:
  print("It is clear") 
```

## Loops

* ## Loop while
The while loop in Python is a flow control statement used to execute a block of instructions repeatedly as long as a given condition is met. That is, as long as the while condition is met, a block of code will be executed repeatedly [(J2LOGO)](https://j2logo.com/python/tutorial/python-while-bucle/).


   * ## Example 1
We use a print() and a while that will count progressively until the number 6 is reached, once it stops and its condition is fulfilled. 

```python
print("counter from 1 to 6")
i = 1
while i <= 6:
  print(i)
  i += 1
```

* ## Example 2
We use a print(), a while and an if which at the time that the while loop reaches its arrival condition the if given that it is equal to 2 when it passes through there the while loop will skip it.

```python
print("jumping counter ")
i = 0
while i < 6:
  i += 1
  if i == 2:
    continue
  print(i)
```

* ## Example 3
We use a print() and a while that will count down starting from the number 10 to 0 once it arrives it stops and its condition is fulfilled.

```python
print("Countdown counter")
i = 10
while i > 0:
  print(i)
  i -= 1
```

* ## Loop for
The for loop is used to repeat elements of an iterable object (list, tuple, set, dictionary, ...) with a predetermined number of times [(J2LOGO)](https://j2logo.com/bucle-for-en-python/).

   * ### Example 1
   We use a print() and a for that will count progressively until the number 9, once it is reached, it stops. 

```python
print("counter from 1 to 9")

for i in range(9):
    print(i+1)
```

* ## Example 2
We use a print(), a for that will count quadruple and another print() which until it reaches the number 16 will stop printing numbers.

```python
print("4 in 4 counter")
for i in range (0, 20, 4):
 print(i)
```

* ## Example 3
A print() is used, a string of numbers is stored in a variable, which the fort loop will pass one by one until its arithmetic sentence is fulfilled and an even number is found and the last row is reached. 

```python
print ("collector of even numbers")
coleccion = [2, 4, 5, 7, 8, 9, 3, 4]
for i in coleccion:
    if i % 2 != 0:
        continue
    print(i)
```
